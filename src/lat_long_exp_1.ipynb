```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style("darkgrid")
from sklearn import linear_model
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import random
import matplotlib.patches as mpatches

import warnings
warnings.filterwarnings('ignore')
```

```python
# Levanto los datos


def crear_anio(datos,tag):
    return datos[tag].apply (lambda x : x.year if x.month < 12 else x.year + 1)
def crear_mes(datos,tag):
    return datos[tag].apply (lambda x : x.month)

data_temperatura = pd.read_csv('../data/GlobalLandTemperaturesByCity.csv', parse_dates=['dt'])

data_lat_long = pd.read_csv('../data/long_lat/cities_lat_long.csv')

df = pd.merge(data_temperatura, data_lat_long, on=['City', 'Country'])


#df = df.dropna(axis=1, how='any')
#df = df.drop('iso2',1)
#df = df.drop('iso3', 1)
#df = df.drop('pop', 1)
#df = df.drop('Latitude', 1)
#df = df.drop('Longitude', 1)

```

```python
df_testing = df

df_testing['month'] = crear_mes(df_testing, 'dt')
df_testing['year'] = crear_anio(df_testing, 'dt')
```

```python
# GrÃ¡ficos los datos

df_new = df_testing[(df_testing['lng'] < -65) & (df_testing['lng'] > -75) & (df_testing['lat'] < 40) & (df_testing['lat'] > -50)]


df_new = df_new[(df_new['month'] == 12) & (df_new['year'] == 2000)]

Oruro = df_new[df_new['City'] == 'Oruro']
Santiago = df_new[df_new['City'] == 'Santiago']
Maracaibo = df_new[df_new['City'] == 'Maracaibo']

grouped = df_new[(df_new['City'] != 'oruro') & (df_new['City'] != 'Santiago') & (df_new['City'] != 'Maracaibo')]

#grouped['City']

grouped = grouped.groupby(['City']).min()

Oruro = {
    "avg": Oruro['AverageTemperature'].iloc[0],
    "lat": Oruro['lat'].iloc[0]
}

Maracaibo = {
    "avg": Maracaibo['AverageTemperature'].iloc[0],
    "lat": Maracaibo['lat'].iloc[0]
}

Santiago = {
    "avg": Santiago['AverageTemperature'].iloc[0],
    "lat": Santiago['lat'].iloc[0]
}


#grouped.head(1000)


```

```python
# Uso cuadrados mÃ­nimos para predecir
sns.tsplot(data=grouped['AverageTemperature'], time=grouped['lat'], interpolate=False, color='b')


#Santiago_legend = mpatches.Patch(color='red', label='The red data')
#plt.legend(handles=[Santiago_legend])

#plt.plot([3,2,1], label="test2")
# Place a legend to the right of this smaller subplot.

font = {'family': 'normal', 'weight': 'bold', 'size': 20}

sns.plt.rc('font', **font)

sns.plt.rc('axes', labelsize=14)


ax_1 = sns.plt.plot((Santiago['lat'], Santiago['lat']), (Santiago['avg'], Santiago['avg']), 'b-o', label= 'Santiago', marker='o', color='g')
sns.plt.plot((Maracaibo['lat'], Maracaibo['lat']), (Maracaibo['avg'], Maracaibo['avg']), 'b-o', label= 'Maracaibo', marker='o' , color='y')
sns.plt.plot((Oruro['lat'], Oruro['lat']), (Oruro['avg'], Oruro['avg']), 'b-o', label= 'Oruro', marker='o', color='purple')

plt.legend(loc='best', numpoints=1, handlelength=0)

#plt.title('Promedio de temperaturas de diciembre en distintas latitudes')

def armar_matriz_A(s):
    temp = []
    for t in s:
        temp.append([t**3, t, 1, np.sin(0.8*t)*np.cos(t)]) #3
        #temp.append([t**2, t, 1, np.sin(t)]) #2
        #temp.append([t**2, t, 1]) #1
    return np.array(temp)

def entrenar_y_predecir_en_rangos(df, min_rango_entrenamiento, max_rango_entrenamiento):
    regr = linear_model.LinearRegression(fit_intercept=False)
    
    
    # Entreno el modelo
    df_e = df[(df['lat'] <= max_rango_entrenamiento) & (df['lat'] >= min_rango_entrenamiento)]
    A_e = armar_matriz_A(df_e['lat'])
    regr.fit(A_e, df_e['AverageTemperature'])

    # Me fijo la aproximaciÃ³n que se realizÃ³
    df_e['p'] = regr.predict(A_e)
    df_e['Latitud'] = df_e['lat']
    sns.tsplot(time=df_e['Latitud'], data=df_e['p'], color='r')
    return 1

def predecir(k):
    return entrenar_y_predecir_en_rangos(grouped, k-100, k)
    
plt.ylabel('AVG Temperatura')
ECM = predecir(50)
sns.plt.show()
```
